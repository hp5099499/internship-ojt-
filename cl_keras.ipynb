```python
# Classification is a type of machine learning algorithm used to predict a categorical label.
# Import required libraries
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import sklearn

# Import necessary modules
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from math import sqrt

# Keras specific
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import to_categorical
```

```python
df=pd.read_csv('diabetes.csv')
print(df.shape)
df.describe()
```

```python
# Creating Arrays for the Features and the Response Variable.
target_column = ['Outcome'] 
predictors=list(set(list(df.columns))-set(target_column))
df[predictors]=df[predictors]/df[predictors].max()
df.describe()
```

```python
X=df[predictors].values
y = df[target_column].values
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=40)
X_train.shape,X_test.shape
```

```python
# one hot encode outputs
y_train=to_categorical(y_train)
y_test=to_categorical(y_test)
count_classes=y_test.shape[1]
print(count_classes)
```

```python
#  Define, Compile, and Fit the Keras Classification Model
model=Sequential()
model.add(Dense(500,activation='relu',input_dim=8))
model.add(Dense(100,activation='relu'))
model.add(Dense(50,activation='relu'))
model.add(Dense(2,activation='softmax'))
```

```python
# compile the model
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
model.fit(X_train,y_train,epochs=20)
```

```python
# Predict on the Test Data and Compute Evaluation Metrics;
pred_train=model(X_train)
score=model.evaluate(X_train,y_train,verbose=0)
print('Accuracy on training data:{}% \n Error on training data:{}'.format(score[1],1-score[1]))

pred_test=model(X_test)
score=model.evaluate(X_test,y_test,verbose=0)
print('Accuracy on training data:{}% \n Error on training data:{}'.format(score[1],1-score[1]))
```

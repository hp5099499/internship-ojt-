```python
sns.scatterplot(x="total_bill" , y="tip" , data=df , hue="time")
```

```python
import matplotlib.piplot as plt
sns.relplot(
    data=df,
    x = "total_bill", y = "tip",
    hue ="time",
    style = "time",
    sizes =(10,100) , palette = "hot_r"
)
plt.xlabel("Total Bill In USD $")
plt.ylabel("Tips")
plt.title("Total Bills Vs Tips")
```

```python
import pandas as pd
import numpy as np
import cufflinks as cf
import seaborn as sns
import plotly.express as px
%matplotlib inline
from plotly.offline import download_plotlyjs, init_notebook_mode,plot,iplot
init_notebook_mode(connected=True)
cf.go_offline()
```

```python
# line chart
import plotly.graph_objects as go
df_stocks=px.data.stocks()
px.line(df_stocks,x='date',y='GOOG',labels={'x':'Date','y':'Price'})
px.line(df_stocks,x='date',y=['GOOG','AAPL'],labels='Apple vs Google')
fig=go.Figure()
fig.add_trace(go.Scatter(x=df_stocks.date,y=df_stocks.AAPL,mode='lines+markers',name="APPLE"))
fig.add_trace(go.Scatter(x=df_stocks.date,y=df_stocks.AMZN,mode='lines+markers',name="Amzn"))
fig.add_trace(go.Scatter(x=df_stocks.date,y=df_stocks.GOOG,mode='lines+markers',name="GOOGLE",line=dict(color='green',width=2,dash='dashdot')))
# fig.update_layout(title='Stock price data 2018-2020',xaxis_title='Price',yaxis_title='Date')
fig.update_layout(
    xaxis=dict(
        showline=True,
        showgrid=False,
        showticklabels=True,
        linecolor='rgb(42, 45, 37)',
        linewidth=2,
        ticks='outside',
        tickfont=dict(
            family='Arial',
            size=12,
            color='rgb(82, 82, 82)'
        )
    )
)

fig.update_layout(
    xaxis=dict(
        showline=True,
        showgrid=False,
        showticklabels=True,
        linecolor='rgb(42, 45, 37)',
        linewidth=2,
        ticks='outside',
        tickfont=dict(
            family='Arial',
            size=12,
            color='rgb(82, 82, 82)'
        )
    ),
    yaxis=dict(
        showgrid=False,
        zeroline=False,
        showticklabels=False,
        showline=False
    ),
    autosize=False,
    margin=dict(
        autoexpand=False,
        l=100,
        r=20,
        t=110
    ),
    showlegend=False,
    plot_bgcolor='white'
)

```

```python
#barchart
df_us=px.data.gapminder().query("country=='United States'")
px.bar(df_us,x='year',y='pop')

df_tips=px.data.tips()
px.bar(df_tips,x='day',y='tip',color='sex',title="Tips by sex on each day",
      labels={'tip':'Tips on each Day','day':'Day of the Week'})
px.bar(df_tips,x='day',y='tip',color='smoker',barmode='group')


df_europe = px.data.gapminder().query("continent=='Europe' and year==2007 and pop > 2.e6")
fig = px.bar(df_europe, y='pop', x='country', text='pop', color='country')
fig.update_traces(texttemplate='%{text:.2s}', textposition='outside')
fig.update_layout(
    uniformtext_minsize=8,
    xaxis_tickangle=-45
)

fig.show()
```

```python
# scatter plot
import plotly.express as px

df_iris = px.data.iris()
px.scatter(
    df_iris,
    x='sepal_width',
    size='petal_length',  # Only specify the size parameter once
    color='species',
    hover_data=['petal_width']
)
fig=go.Figure()

fig.add_trace(go.Scatter(
    x=df_iris.sepal_width,
    y=df_iris.sepal_length,
    mode='markers',
    marker=dict(
        color=df_iris.sepal_width,
        showscale=True,
        line=dict(width=2),
        size=10
    ),
    text=df_iris.species
))

# Update the marker properties
fig.update_traces(marker_line_width=2, marker_size=10)

fig = go.Figure(data=go.Scattergl(
    x=np.random.randn(10000),
    y=np.random.randn(10000),
    mode='markers',
    marker=dict(
        color=np.random.randn(10000),
        colorscale='Viridis',
        line_width=1
    )
))
fig

```

```python
#pie charts
df_asia=px.data.gapminder().query("year==2007").query("continent=='Asia'")
px.pie(df_asia,values='pop',names='country',
                              title="Population of Asian Continent",
                              color_discrete_sequence=px.colors.sequential.RdBu)
colors=['blue','red','black','green','brown','purple']
fig=go.Figure(data=[go.Pie(labels=['water','Grass','Normal','Physic','fire','Ground'],
                          values=[110,130,143,290,234,144])])
fig.update_traces(hoverinfo='label+percent',textfont_size=20,
                  textinfo='label+percent',
                 pull=[0.1,0,0.2,0,0,0],marker=dict(colors=colors,line=dict(color='#FFFFFF',width=2)))
```

```python
# Histogram
dice1=np.random.randint(1,7,5000)
dice2=np.random.randint(1,7,5000)
dice_sum=dice1+dice2
fig=px.histogram(dice_sum,nbins=11,
                labels={'value':'Dice Roll'},
                title='5000 Dice Roll Histogram',
                marginal='violin',
                color_discrete_sequence=['green'])
fig.update_layout(
xaxis_title_text='Dice Roll',
yaxis_title_text='Dice sum',
    bargap=0.2,showlegend=False)
df_tips=px.data.tips()
px.histogram(df_tips,x='total_bill',color='sex')
```

```python
# Boxplot
df_tips=px.data.tips()
px.box(df_tips,x='sex',y='tip',points='all')
```

```python
px.box(df_tips,x='day',y='tip',color='sex')
fig=go.Figure()
fig.add_trace(go.Box(x=df_tips.sex,y=df_tips.tip,marker_color='blue',boxmean='sd'))
```

```python
df_stocks=px.data.stocks()
fig=go.Figure()
fig.add_trace(go.Box(y=df_stocks.GOOG,boxpoints='all',fillcolor='blue',jitter=0.5,whiskerwidth=0.2))
fig.add_trace(go.Box(y=df_stocks.AAPL,boxpoints='all',fillcolor='green',jitter=0.5,whiskerwidth=0.2))
fig.update_layout(title='Google vs Apple',
                 yaxis=dict(gridcolor='rgb(255,255,255)',
                           gridwidth=3),
                paper_bgcolor='rgb(243,243,243)',
                plot_bgcolor='rgb(243,243,243)' )                 
```

```python
# violinplot
px.violin(df_tips,y='total_bill',box=True,points='all')
```

```python
px.violin(df_tips,y='tip',x='smoker',color='sex',box=True,
         points='all',hover_data=df_tips.columns)
```

```python
fig=go.Figure()

fig.add_trace(go.Violin(x=df_tips['day'][df_tips['smoker']=='Yes'],y=df_tips['total_bill'][df_tips['smoker']=='Yes'],legendgroup='Yes',
                       scalegroup='Yes',name='Yes',side='negative',line_color='blue'))
fig.add_trace(go.Violin(x=df_tips['day'][df_tips['smoker']=='No'],y=df_tips['total_bill'][df_tips['smoker']=='No'],legendgroup='Yes',
                       scalegroup='Yes',name='No',side='positive',line_color='red'))
```

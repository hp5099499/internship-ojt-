```python
import pandas as pd
df = pd.read_csv('class31.csv')
df
```

```python
#lets make the function for revenue
def treven():
    df['Total_revenue'] = df['Price']*df['quantity sold']
    return df
treven()
```

```python
#lets make another function for extract
def textract():
    df['username'] = df['email'].str.split('@').str[0]
    return df
textract()
```

```python
#lets make another function for calculating final price
def final_price():
    df['dis_amount'] = df['Price'] * (df['discount%']/100)
    df['final_price'] = df['Price'] - df['dis_amount']
    return df
final_price()
```

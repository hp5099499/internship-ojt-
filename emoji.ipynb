```python
f=open('glove.6B.100d.txt','r',encoding='utf8')
c=f.readlines()

# f.close()
```

```python
embedding={}
```

```python
import numpy as np
for line in c:
    line=line.split()
    embedding[line[0]]=np.array(line[1:],dtype=float)
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.utils import to_categorical
```

```python
tokenizer=Tokenizer()
```

```python
tokenizer.fit_on_texts(x)
w2i=tokenizer.word_index
xtokens=tokenizer.texts_to_sequences
```

```python
xtokens=tokenizer.texts_to_sequences(x)
xtokens
```

```python
w2i
```

```python
def get_max(df):
    max_len=0
    for sent in df:
        max_len=max(max_len,len(sent))
    return max_len
```

```python
max_len=get_max(xtokens)
max_len
```

```python
xtrain=pad_sequences(xtokens,maxlen=max_len,padding='post')
xtokens[:3]
```

```python
xtrain[:3]
```

```python
embed_s=100
e_matrix=np.zeros((len(w2i)+1,embed_s))
```

```python
e_matrix[1]
```

```python
for word,a in w2i.items():
    e_vector=embedding[word]
    e_matrix[a]=e_vector
```

```python
e_matrix[1]
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Embedding, SimpleRNN
model=Sequential([Embedding(input_dim=len(w2i)+1,
                           output_dim=embed_s,
                           input_length=max_len,
                           weights=[e_matrix]),
                 LSTM(units=16,return_sequences=True),
                 LSTM(units=4),
                 Dense(5,activation='sigmoid')])
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
y
```

```python

ytrain = to_categorical(y)
ytrain
# print(ytrain.shape,xtrain.shape)

```

```python
model.fit(xtrain, ytrain, epochs=10)
```

```python
model.summary()
```

```python

test = ['i am hungry', 'playing', 'in love', 'valentine week']
test_s = tokenizer.texts_to_sequences(test)
xtest = pad_sequences(test_s, maxlen=max_len, padding = 'post')

y_pred = model.predict(xtest)
```

```python
y_pred
```

```python
y_pred = np.argmax(y_pred, axis=1)

y_pred
```
